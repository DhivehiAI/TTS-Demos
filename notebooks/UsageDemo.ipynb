### Initialize the Model for Inference
We need to provide the model and vocoder paths to TTSModel

```python
import os
from IPython.display import Audio
from tts import TTSModel

model_path = "./Male Goveli"
tts = TTSModel(
    os.path.join(model_path, "tts.saved_model"),
    os.path.join(model_path, "vocoder.saved_model")
)
```

### Run Inference

```python
# actual inference to get raw Wav
sample_text = "އަހަރެންނަކީ ދިވެހިބަހުން ވާހަކަދައްކާ، މަސްނޫވީ އެއްޗެއް"
result = tts(sample_text, speed=0.95)
```

```python
# display results
Audio(result, rate=22050)
```
